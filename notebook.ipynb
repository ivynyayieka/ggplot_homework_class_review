I think the KDE plots out the points of the data. When there are many points on one specific location on the data, the estimation is higher and the possibility of seeing a point on the KDE curve increases. The kernel function is the mathematical way to calculate the points on the curve. 

### Box and Whisker Plot
http://www.sthda.com/english/wiki/ggplot2-box-plot-quick-start-guide-r-software-and-data-visualization



````python
%%R -w 1000 -h 200

ggplot(df) + 
    aes(x=pct_below_poverty, y=1) +
    geom_boxplot(alpha=.2,outlier.size=10) +
      theme(axis.title.y=element_blank(),
            axis.text.y=element_blank(),  #remove y axis labels
            axis.ticks.y=element_blank()  #remove y axis ticks
            )+ 
    labs(title = "Box plot of percentage of people below poverty",
              subtitle = "In most NYC community districts, about 10% to 20% of people are living below poverty with 3 significant outliers on the higher end",
              caption = "Data source: homework") +
    xlab("% of population below poverty")
````

Box and whisker plots aren't super easy to read. We don't always use them in reader-facing visualizations, but you'll find them a lot in scientific papers. Box and whisker plots help draw attention to outliers in relation to the median and interquartile range of the data.

````python
%%R -w 1000 -h 200

#Define function to find the outlier

find_outlier <- function(x) {
  return(x < quantile(x, .25) - 1.5*IQR(x) | x > quantile(x, .75) + 1.5*IQR(x))
}
# modify this code

#call necessary libraries?
library(ggplot2)
library(dplyr)

#add new column to data frame that indicates if each observation is an outlier
df <- df %>%
        mutate(outlier = ifelse(find_outlier(pct_below_poverty), zip, NA))


ggplot(df) + 
    aes(x=pct_below_poverty, y=1) +
    geom_boxplot(alpha=.2,outlier.size=10) +
     geom_text(aes(label=outlier), na.rm=TRUE, hjust=-.5)
````
